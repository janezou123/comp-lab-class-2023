```python
import numpy as np
import matplotlib.pyplot as plt

# Define a function to read XVG files
def read_xvg(file_path):
    data = {"x": [], "y": []}
    with open(file_path, 'r') as f:
        for line in f:
            if line.startswith(('#', '@')):
                continue
            parts = line.strip().split()
            if len(parts) == 2:
                x, y = map(float, parts)
                data["x"].append(x)
                data["y"].append(y)
    return data


```

```python
# Specify the path to your XVG file
xvg_file_path = "/home/hz2750/comp-lab-class/comp-lab-class-2023/Week3/Analysis/potential.xvg"

# Read data from the XVG file
data = read_xvg(xvg_file_path)

# Create a plot
plt.figure(figsize=(10, 6))
plt.plot(data["x"], data["y"], label="Data")
plt.xlabel("Energy Minimiazation")
plt.ylabel("Potential Energy (KJ mol^-1)")
plt.title("Energy Minimization")
plt.grid(True)

# Show the plot in the Jupyter Notebook
plt.show()
```

```python
# Define a function to read XVG files with more than 2 columns and plot the first two columns
def read_xvg_multi_and_plot(file_path):
    data = {"x": [], "y": []}
    with open(file_path, 'r') as f:
        for line in f:
            if line.startswith(('#', '@')):
                continue
            parts = line.strip().split()
            if len(parts) >= 2:  # Ensure there are at least 2 columns
                x, y = map(float, parts[:2])  # Take the first two columns
                data["x"].append(x)
                data["y"].append(y)
    return data

# Specify the path to your XVG file
xvg_file_path = "/home/hz2750/comp-lab-class/comp-lab-class-2023/Week3/Analysis/gyrate.xvg"

# Read and plot data from the XVG file (first two columns)
data = read_xvg_multi_and_plot(xvg_file_path)

# Create a plot
plt.figure(figsize=(10, 6))
plt.plot(data["x"], data["y"], label="data")
plt.xlabel("Time(ps)")
plt.ylabel("Rg (nm)")
plt.title("Radius of gyration (total and around axes)")
plt.grid(True)

# Show the plot in the Jupyter Notebook
plt.show()
```
