```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Part2
#loadfile
file_energy = np.loadtxt('/home/hz2750/comp-lab-class/comp-lab-class-2023/Week13/Outputs/NVT-1.ener')

#Dataframe
column_energy = pd.DataFrame(file_energy, columns=['Step Nr.', 'Time[fs]', 'Kin.[a.u.]', 'Temp[K]', 'Pot.[a.u.]', 'Cons Qty[a.u.]', 'UsedTime[s]'])
```

```python
# Plot with more control over aesthetics
plt.plot(column_energy['Time[fs]'], column_energy['Pot.[a.u.]'])

# Setting axis labels and plot title
plt.xlabel('Time (fs)', fontsize=12)
plt.ylabel('Potential Energy (a.u.)', fontsize=12)
plt.title('Potential Energy vs Time_NVT-1', fontsize=14)

# Optionally, set specific limits for the axes if needed (uncomment and adjust if needed)
# plt.xlim(min_value, max_value)
# plt.ylim(min_value, max_value)

# Improve the layout and grid visibility
plt.grid(True)
plt.tight_layout()
plt.savefig('/home/hz2750/comp-lab-class/comp-lab-class-2023/Week13/Figures/Potential Energy vs Time_NVT-1')
# Save the plot with a higher resolution
#plt.savefig('/home/hz2750/comp-lab-class/comp-lab-class-2023/Week12/Figures/potential_E_versus_time_water.png', dpi=300)

# Display the plot
plt.show()
```
