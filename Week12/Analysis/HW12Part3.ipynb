```python
import MDAnalysis as mda
from MDAnalysis.analysis.rdf import InterRDF
import matplotlib.pyplot as plt
import numpy as np

# Load the trajectory and topology
# Replace these file names with your actual files
u = mda.Universe('/home/hz2750/comp-lab-class/comp-lab-class-2023/Week12/Runs/Proton/proton_aimd_1ns_every0.5fs-pos-1.xyz')

# Set the unit cell dimensions - 12.42  converted to nanometers
unit_cell_length = 12.42 / 10  # Convert to nanometers
for ts in u.trajectory:
    ts.dimensions = [unit_cell_length, unit_cell_length, unit_cell_length, 90, 90, 90]

# Select oxygen atoms
oxygen_atoms = u.select_atoms('name O')

# Compute RDF
# Adjust range and number of bins as needed
r_range = (0.0, 1.5)  # Range in nanometers
rdf = InterRDF(oxygen_atoms, oxygen_atoms, range=r_range, nbins=100)
rdf.run()

# Plotting
plt.plot(rdf.bins, rdf.rdf)
plt.xlabel('r (nm)')
plt.ylabel('g(r)')
plt.title('Radial Distribution Function of Oxygen Atoms_Proton')
plt.savefig('/home/hz2750/comp-lab-class/comp-lab-class-2023/Week12/Figures/Radial Distribution Function of Oxygen Atoms_Proton.png')
plt.show()
```
