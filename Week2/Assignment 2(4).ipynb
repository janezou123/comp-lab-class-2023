```python
#b) Make 6 separate plots for each quantity vs time, with time in ns
import matplotlib.pyplot as plt
import numpy as np

# Initialize empty lists to store data for each quantity
time_ns = []
potential_data = []
kinetic_energy_data = []
total_energy_data = []
temperature_data = []
pressure_data = []
volume_data = []

# Read the XVG file (replace 'your_data_file.xvg' with your actual file path)
with open("Data/1hz3_T310.run.25000000.energy.xvg", 'r') as file:
    lines = file.readlines()
    for line in lines[24:]:
        values = line.split()
        if len(values) == 7:
            time_ns.append(float(values[0]) / 1000.0)  # Convert ps to ns
            potential_data.append(float(values[1]))
            kinetic_energy_data.append(float(values[2]))
            total_energy_data.append(float(values[3]))
            temperature_data.append(float(values[4]))
            pressure_data.append(float(values[5]))
            volume_data.append(float(values[6]))

# Create six separate plots for each quantity vs. time
quantities = [
    ("Potential", potential_data),
    ("Kinetic En.", kinetic_energy_data),
    ("Total Energy", total_energy_data),
    ("Temperature", temperature_data),
    ("Pressure", pressure_data),
    ("Volume", volume_data)
]

for quantity_name, quantity_data in quantities:
    plt.figure()
    plt.plot(time_ns, quantity_data)
    plt.xlabel("Time (ns)")
    plt.ylabel(quantity_name)
    plt.title(f"{quantity_name} vs. Time (ns)")
    plt.grid()

# Display all the plots
plt.show()







```

```python
#c) Make a plot that has kinetic energy and temperature on the same plot, where the left axis is temperature and the right axis is kinetic energy, using different colors and different axis labels
import matplotlib.pyplot as plt
import numpy as np

# Load data from the XVG file (replace 'your_data_file.xvg' with your actual file path)
data = []
with open("Data/1hz3_T310.run.25000000.energy.xvg", 'r') as file:
    for line in file:
        if line.startswith(('@', '#', '&')):
            continue  # Skip header lines
        parts = line.split()
        if len(parts) >= 7:
            data.append([float(parts[0]), float(parts[1]), float(parts[2]), float(parts[3]), float(parts[4]), float(parts[5]), float(parts[6])])

data = np.array(data)
time_ps = data[:, 0]
kinetic_energy_data = data[:, 1]
temperature_data = data[:, 3]

# Convert time from picoseconds (ps) to nanoseconds (ns)
time_ns = time_ps * 0.001  # 1 ps = 0.001 ns

# Create a figure and axis for the plot
fig, ax1 = plt.subplots()

# Plot temperature data on the left y-axis (ax1)
color = 'tab:red'
ax1.set_xlabel("Time (ns)")
ax1.set_ylabel("Temperature (K)", color=color)
ax1.plot(time_ns, temperature_data, color=color, label="Temperature")
ax1.tick_params(axis='y', labelcolor=color)

# Create a second y-axis on the right for kinetic energy
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel("Kinetic Energy (kJ/mol)", color=color)
ax2.plot(time_ns, kinetic_energy_data, color=color, label="Kinetic Energy")
ax2.tick_params(axis='y', labelcolor=color)

# Add a legend for both lines
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines = lines1 + lines2
labels = labels1 + labels2
ax1.legend(lines, labels, loc="upper right")

# Add a title to the plot
plt.title("Kinetic Energy and Temperature vs. Time")

# Display the plot
plt.grid()
plt.tight_layout()
plt.show()


```
