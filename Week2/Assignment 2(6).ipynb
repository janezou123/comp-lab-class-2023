```python
import mdtraj as md
import numpy as np
import matplotlib.pyplot as plt
```

```python
# a) Read in the MD trajectory of beta amyloid
traj = md.load('/scratch/work/courses/CHEM-GA-2671-2023fa/students/hz2750/comp-lab-class-2023/Week2/Data/1hz3_T310.stepid25000000.every100ps.nowater.xtc', top='/scratch/work/courses/CHEM-GA-2671-2023fa/students/hz2750/comp-lab-class-2023/Week2/Data/1hz3_T310.start.nowater.gro')
end_to_end_distances = md.compute_distances(traj, [[0, len(traj) - 1]])
radii_of_gyration = md.compute_rg(traj)
```

```python
# b) Compute the end-end distance and radius of gyration at each time
end_to_end_distances = md.compute_distances(traj, [[0, len(traj) - 1]])
radii_of_gyration = md.compute_rg(traj)
```

```python
# c) Plot the end-end distance and radius of gyration vs time
plt.figure(figsize=(10, 6))
plt.plot(traj.time, end_to_end_distances, label='End-to-End Distance')
plt.plot(traj.time, radii_of_gyration, label='Radius of Gyration')
plt.xlabel('Time (ps)')
plt.ylabel('Distance (angstroms)')
plt.legend()
plt.title('End-to-End Distance and Radius of Gyration vs Time')
plt.grid(True)
```

```python
# d) Plot normalized histograms
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(end_to_end_distances, bins=50, density=True, alpha=0.7)
plt.xlabel('End-to-End Distance (angstroms)')
plt.ylabel('Normalized Frequency')
plt.title('Normalized Histogram of End-to-End Distance')

plt.subplot(1, 2, 2)
plt.hist(radii_of_gyration, bins=50, density=True, alpha=0.7)
plt.xlabel('Radius of Gyration (angstroms)')
plt.ylabel('Normalized Frequency')
plt.title('Normalized Histogram of Radius of Gyration')

plt.tight_layout()
```

```python
# Show all the plots
plt.show()
```
