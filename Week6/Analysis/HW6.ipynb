```python
#Exercise 1-TopolA
import matplotlib.pyplot as plt

# Lists to store time and CV values
time_ps = []
cv_radians = []

# Read data from the COLVAR file
with open('/home/hz2750/comp-lab-class/comp-lab-class-2023/Week6/Q3/topolA/COLVAR', 'r') as colvar_file:
    for line in colvar_file:
        if line.strip() and not line.startswith("#"):
            data = line.split()
            time_ps.append(float(data[0]))  # Assuming time is in the first column
            cv_radians.append(float(data[1]))  # Assuming CV in radians is in the second column

# Convert time from ps to ns
time_ns = [t / 1000 for t in time_ps]

# Create the plot
plt.figure(figsize=(10, 5))
plt.plot(time_ns, cv_radians, label='Metadynamics CV')
plt.title('Time evolution of the metadynamics CV during the first 2 ns of a metadynamics simulation of alanine dipeptide in vacuum_TopolA')
plt.xlabel('Time (ns)')
plt.ylabel('Radians')
plt.grid(True)
plt.legend()
plt.xlim(0, 2)  # Set the x-axis limits to 0-2 ns
plt.show()
plt.savefig("/home/hz2750/comp-lab-class/comp-lab-class-2023/Week6/Figures/2ns_TopolA.png")
```

```python
#2D surface map
import numpy as np
import matplotlib.pyplot as plt

# Load your data from 'fes.dat'
data = np.loadtxt('/home/hz2750/comp-lab-class/comp-lab-class-2023/Week6/Q6/fes.dat')

# Extract x (phi), y (psi), and free energy values
phi = data[:, 0]
psi = data[:, 1]
free_energy = data[:, 3]

# Set up the figure and axis
fig, ax = plt.subplots()

# Create a meshgrid for x (phi) and y (psi)
phi_values = np.unique(phi)
psi_values = np.unique(psi)
X, Y = np.meshgrid(phi_values, psi_values)

# Plot the FES using pcolormesh with the 'viridis' colormap
pcm = ax.pcolormesh(X, Y, free_energy.reshape(len(psi_values), len(phi_values)), cmap='viridis', shading='auto')

# Add a color bar with a label
cbar = fig.colorbar(pcm, ax=ax, label='Free Energy')

# Add contour lines with smoothed interpolation
contours = ax.contour(X, Y, free_energy.reshape(len(psi_values), len(phi_values)), levels=10, colors='black', linestyles='solid')

# Label the axes with appropriate labels
ax.set_xlabel('Phi')
ax.set_ylabel('Psi')

# Show

plt.savefig("/home/hz2750/comp-lab-class/comp-lab-class-2023/Week6/Figures/2D_Q6.png")


```

```python
#bonus
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from matplotlib.animation import FuncAnimation

# Load your data from 'fes.dat'
data = np.loadtxt('/home/hz2750/comp-lab-class/comp-lab-class-2023/Week6/Q6/fes.dat')

# Extract x (phi), y (psi), and free energy values
phi = data[:, 0]
psi = data[:, 1]
free_energy = data[:, 3]

# Set up the figure and axis
fig, ax = plt.subplots()

# Create a meshgrid for x (phi) and y (psi)
phi_values = np.unique(phi)
psi_values = np.unique(psi)
X, Y = np.meshgrid(phi_values, psi_values)

# Create a custom colormap transitioning from blue to yellow to red
colors = ['#0000ff', '#00ffff', '#ffff00', '#ff0000']
custom_cmap = ListedColormap(colors)

# Function to update the animation
def update(frame):
    ax.clear()
    surface_data = free_energy[frame]
    pcm = ax.pcolormesh(X, Y, surface_data.reshape(len(psi_values), len(phi_values)), cmap=custom_cmap, shading='auto')
    ax.contour(X, Y, surface_data.reshape(len(psi_values), len(phi_values)), levels=10, colors='black', linestyles='solid')
    ax.set_xlabel('Phi')
    ax.set_ylabel('Psi')
    cbar = fig.colorbar(pcm, ax=ax, label='Free Energy')
    return cbar

# Create an animation by iterating through the surfaces
frames = [i for i in range(data.shape[0])]

animation = FuncAnimation(fig, update, frames=frames, repeat=False, blit=True)

# Save the animation as a GIF
animation.save('fes_animation.gif', writer='pillow', fps=10)

plt.show()

```
