```python
import numpy as np
import matplotlib.pyplot as plt

# Define the paths to the XVG files
file_paths = {
    'T300': '/home/hz2750/comp-lab-class/comp-lab-class-2023/Week5/Directories/T300/T300_energy.xvg',
    'T363': '/home/hz2750/comp-lab-class/comp-lab-class-2023/Week5/Directories/T363/T363_energy.xvg',
    'T440': '/home/hz2750/comp-lab-class/comp-lab-class-2023/Week5/Directories/T440/T440_energy.xvg'
}

# Load potential energy data for each temperature
energy_data = []

for temp, file_path in file_paths.items():
    data = np.genfromtxt(file_path, skip_header=25)
    energy_data.append(data[:, 1])

# Compute histograms for each dataset
histograms = []

for energies in energy_data:
    hist, bin_edges = np.histogram(energies, bins=100, density=True)
    histograms.append(hist)

# Plot all histograms on the same plot
for temp, hist in zip(file_paths.keys(), histograms):
    plt.plot(bin_edges[:-1], hist, label=temp)

plt.xlabel('Potential Energy')
plt.ylabel('Density')
plt.legend()
plt.title('Potential Energy Histograms for Different Temperatures')
plt.savefig("/home/hz2750/comp-lab-class/comp-lab-class-2023/Week5/Figures/PotentialEvsDifferentT.png")
plt.show()

```
